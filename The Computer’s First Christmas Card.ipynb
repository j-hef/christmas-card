```python
from random import choice, randint

holiday_words = ["jolly", "merry", "holly", "berry", "happy"] 
names = ["molly", "jerry", "harry", "barry", "jarry"]
christmas_list = list("christmas")
chris = "Christmas"[0:5]
chr_ = ["chrysanthemum", "chromoplast", "chronicle", "chronograph", "chromosome", "chromaticity", "chromium", "chromaticscale"]

first_letters = list(set(word[0] for word in holiday_words))
vowels = list(set(word[1] for word in holiday_words))
endings = list(set(word[2:] for word in holiday_words))

new_holiday_words = []
for letter in first_letters:
    for vowel in vowels:
        for ending in endings:
            new_word = letter + vowel + ending
            new_holiday_words.append(new_word)

# new_holiday_words.remove("helly")
# new_holiday_words.remove("horry")
# new_holiday_words.remove("jeppy")
# new_holiday_words.remove("jappy")
# new_holiday_words.remove("mally")
# new_holiday_words.remove("borry")

def christmas_card():
    card = []

    #first 5 lines
    for i in range(5):
        word_list = holiday_words.copy()
        first_word = choice(word_list)
        word_list.remove(first_word)
        second_word = choice(word_list)
        card.append(first_word + second_word)

    #lines 6-27
    all_words = holiday_words + new_holiday_words
    for i in range(22):
        word_list = all_words.copy()
        first_word = choice(word_list)
        word_list.remove(first_word)
        second_word = choice(word_list)
        if first_word in names:
            first_word = first_word.capitalize()
        if second_word in names: 
            second_word = second_word.capitalize()
        card.append(first_word + second_word)

    #lines 28-35
    card.append(card[0])
    glitch = card[-1][5:]
    for i in range(2):
        card.append(glitch * 2)
    card.append(glitch + chris)
    glitch_line = ""
    j = randint(0,5)
    for i in range(j):
        glitch_line += choice(christmas_list)
    glitch_line += glitch
    if len(glitch_line) < 10:
        for i in range(10 - len(glitch_line)):
            glitch_line += choice(christmas_list)
    card.append(glitch_line)
    card.append(chris + glitch)
    greeting = glitch + choice(chr_)
    greeting = greeting.upper()
    filler_line = ""
    greeting_overflow = greeting[:-10]
    for i in range(10 - len(greeting_overflow)):
        filler_line += choice(christmas_list)
    filler_line += greeting_overflow
    card.append(filler_line)
    card.append(greeting[-10:])
    
    #full card
    return "\n".join(card)
```

```python
print(christmas_card())
```
